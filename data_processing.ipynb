```python
import pandas as pd
df = pd.read_csv('/content/ggdrive/MyDrive/ID08/all_ts_tima_cic_data_28_03_invalid.csv')
df
```

```python
from datetime import time

df = df.replace(r'^\s*$', pd.NA, regex=True)
df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df['CityName'] = df['CityName'].ffill(limit=1)
df['FullNmae'] = df['FullName'].replace('12345',pd.NA)
df['Gender'] = df['Gender'].replace('Male', '0')
df['SoTienDKVayBanDau']= df['SoTienDKVayBanDau'].fillna(df['SoTienDKVayBanDau'].mean())
df['SoTienDKVayBanDau'] = df['SoTienDKVayBanDau'].astype(int)
df['TienGiaiNgan']= df['TienGiaiNgan'].fillna(df['TienGiaiNgan'].mean())
df['TienGiaiNgan'] = df['TienGiaiNgan'].astype(int)
df['application_date'] = pd.to_datetime(df['application_date'], format='%Y-%m-%d', errors='coerce')
df['Name'] = df['Name'].fillna('Không rõ')
df['Address'] = df['Address'].fillna('Không có địa chỉ')
del df['WardName']
del df['ReceiveYourIncomeSalary']
del df['DescriptionPositionJob']
del df['Street']
del df['NameCompany']
del df['DistrictNameCompany']
del df['AddressCompany']
del df['CityCompany']
df['Brieft'] = df['Brieft'].replace('notfound', '0')
df['Brieft'] = df['Brieft'].replace('cb1', '1')
df['Brieft'] = df['Brieft'].replace('cb2', '2')
df['Brieft'] = df['Brieft'].replace('cb3', '3')
df['Brieft'] = df['Brieft'].replace('1cy', '1')
df['Brieft'] = df['Brieft'].replace('2cy', '2')
df['Brieft'] = df['Brieft'].replace('3cy', '3')
df['Brieft'] = df['Brieft'].replace('4cy', '4')
df
```

```python
from scipy.stats import zscore
df['zscore'] = zscore(df['Tiền gốc còn lại'])  # Tính Z-score
outliers_zscore = df[df['zscore'].abs() > 3]    # Phát hiện ngoại lệ
print("Ngoại lệ theo Z-score:")

# Lọc dữ liệu không có ngoại lệ
df1 = df[df['zscore'].abs() <= 3]

df1
```
