# Xử lí dữ liệu thiếu, không hợp lệ hoặc trùng lặp

```python
columns_to_keep = [
    'TS_CREDIT_SCORE_V2', 'SoTienDKVayBanDau', 'TienGiaiNgan', 'SoTienConLai',
    'LongestOverdue', 'NumberOfLoans', 'HasLatePayment', 'HasBadDebt',
    'Gender', 'Birthday', 'Salary', 'Hình thức cư trú', 'Thời gian đã sống',
    'JobName', 'ReceiveYourIncomeSalary', 'ProductCreditName', 'InterestPaymentType',
    'Trạng thái'
]
df = df[columns_to_keep].copy()
df
```

```python
print(df.isnull().sum())
```

```python
import numpy as np

# Xử lý cột Salary (chuyển đổi giá trị không hợp lệ thành NaN)
def clean_salary(salary):
    try:
        # Thay thế các giá trị không hợp lệ
        if isinstance(salary, str):
            salary = salary.replace('$', '').replace('Ten thousand', '10000').strip()
            return float(salary)
        return float(salary)
    except:
        return np.nan

df['Salary'] = df['Salary'].apply(clean_salary)
```

```python
from sklearn.impute import SimpleImputer

# Điền giá trị thiếu bằng giá trị trung bình cho các cột số
numeric_columns = ['SoTienDKVayBanDau', 'TienGiaiNgan', 'Salary']

# Khởi tạo SimpleImputer
imputer_mean = SimpleImputer(strategy='mean')

# Áp dụng imputer cho các cột số
df[numeric_columns] = imputer_mean.fit_transform(df[numeric_columns])

# Kiểm tra lại giá trị thiếu
print(df[numeric_columns].isnull().sum())

df
```

```python
def clean_gender(gender):
    if gender in ['0', '1']:
        return gender
    if isinstance(gender, str):
        if gender.lower() == 'male':
            return 1
        if gender.lower() == 'female':
            return 0
    return np.nan

def clean_birthday(birthday):
    # Chuyển giá trị không phải định dạng ngày hợp lệ thành NaN
    try:
        pd.to_datetime(birthday)
        return birthday
    except:
        return np.nan

def clean_column(value):
    # Chuyển chuỗi rỗng hoặc null thành NaN
    if pd.isna(value) or value == '':
        return np.nan
    return value

# Áp dụng xử lý cho cột Gender
df['Gender'] = df['Gender'].apply(clean_gender)

# Áp dụng xử lý cho cột Birthday
df['Birthday'] = df['Birthday'].apply(clean_birthday)

# Lấy danh sách các cột còn lại (trừ các cột đã xử lý trước đó)
columns_to_clean = [col for col in df.columns if col not in ['SoTienDKVayBanDau', 'TienGiaiNgan', 'Salary', 'Gender', 'Birthday']]

# Áp dụng xử lý cho các cột còn lại
for col in columns_to_clean:
    df[col] = df[col].apply(clean_column)

# Kiểm tra lại giá trị thiếu
print(df.isnull().sum())
```

```python
# Kiểm tra số dòng ban đầu
print(f"Số dòng ban đầu: {len(df)}")

# 1. Kiểm tra và loại bỏ các dòng trùng lặp trên toàn bộ cột
df_no_duplicates_all = df.drop_duplicates(keep='first')

# In số dòng sau khi loại bỏ trùng lặp toàn bộ
print(f"Số dòng sau khi loại bỏ trùng lặp toàn bộ: {len(df_no_duplicates_all)}")
print(f"Số dòng trùng lặp bị xóa: {len(df) - len(df_no_duplicates_all)}")

# 2. Kiểm tra và loại bỏ trùng lặp dựa trên các cột quan trọng
important_columns = [
    'TS_CREDIT_SCORE_V2', 'SoTienDKVayBanDau', 'TienGiaiNgan',
    'Birthday', 'Salary', 'HasBadDebt'
]
df_no_duplicates_subset = df.drop_duplicates(subset=important_columns, keep='first')

# In số dòng sau khi loại bỏ trùng lặp theo cột quan trọng
print(f"Số dòng sau khi loại bỏ trùng lặp theo các cột quan trọng: {len(df_no_duplicates_subset)}")
print(f"Số dòng trùng lặp bị xóa (theo cột quan trọng): {len(df) - len(df_no_duplicates_subset)}")

output_file = 'processed_data_no_duplicates.csv'
df_no_duplicates_subset.to_csv(output_file, index=False)
df = pd.read_csv(output_file)

df
```

# Xử lí dữ liệu ngoại lệ

```python
def detect_outliers_zscore(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    # Tính Z-Score
    df['Z_Score_' + column] = (df[column] - mean) / std
    # Xác định ngoại lệ
    outliers = df[abs(df['Z_Score_' + column]) > threshold]
    return outliers

outliers_summary = []

# Kiểm tra ngoại lệ cho từng cột số
for column in numeric_columns:
    outliers = detect_outliers_zscore(df, column)
    if not outliers.empty:
        print(f"\nNgoại lệ trong cột {column}:")
        print(outliers[[column, 'Z_Score_' + column]])
        outliers_summary.append({
            'Column': column,
            'Number_of_Outliers': len(outliers),
            'Outlier_Values': outliers[column].values
        })

#Thay thế ngoại lệ bằng giá trị trung bình
for column in numeric_columns:
    mean = df[column].mean()
    z_score_column = 'Z_Score_' + column
    df.loc[abs(df[z_score_column]) > 3, column] = mean

# Xóa các cột Z-Score tạm thời
df = df.drop([col for col in df.columns if col.startswith('Z_Score_')], axis=1)

# Lưu DataFrame
df.to_csv('processed_data_no_outliers.csv', index=False)

df
```

# Chuẩn hóa dữ liệu

```python
from sklearn.preprocessing import StandardScaler

# 1. Chuẩn hóa dữ liệu chuỗi
string_columns = [
    'Hình thức cư trú', 'JobName', 'ReceiveYourIncomeSalary',
    'ProductCreditName', 'InterestPaymentType', 'Trạng thái'
]

for col in string_columns:
    # Chuyển thành chuỗi và xử lý giá trị NaN
    df[col] = df[col].astype(str).replace('nan', np.nan)
    # Chuẩn hóa: Chuyển thành chữ hoa đầu từ
    df[col] = df[col].str.title().str.strip()
    # Thay thế các giá trị NaN trở lại
    df[col] = df[col].replace('Nan', np.nan)

print("\nDữ liệu chuỗi sau khi chuẩn hóa:")
for col in string_columns:
    print(f"\nCột {col}:")
    print(df[col].value_counts(dropna=False))

# 2. Chuẩn hóa dữ liệu số
numeric_columns = [
    'TS_CREDIT_SCORE_V2', 'SoTienDKVayBanDau', 'TienGiaiNgan', 'SoTienConLai',
    'LongestOverdue', 'NumberOfLoans', 'Salary'
]

# Xử lý giá trị NaN trong các cột số (điền bằng giá trị trung bình)
for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())

# Khởi tạo StandardScaler
scaler = StandardScaler()

# Chuẩn hóa các cột số
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

print("\nDữ liệu số sau khi chuẩn hóa (StandardScaler):")
print(df[numeric_columns].describe())

# Lưu DataFrame đã chuẩn hóa
df.to_csv('processed_data_normalized.csv', index=False)

df
```

# Tạo các biến mới

```python
from datetime import datetime

df["TyLeGiaiNgan"] = df["TienGiaiNgan"] / df["SoTienDKVayBanDau"]

df["Birthday"] = pd.to_datetime(df["Birthday"], errors='coerce')

df["Age"] = datetime.now().year - df["Birthday"].dt.year

df["TyLeNoSoVoiThuNhap"] = df["SoTienConLai"] / df["Salary"].replace({"Ten thousand": 10000, "$": 23000})
df["TyLeNoSoVoiThuNhap"] = df["TyLeNoSoVoiThuNhap"].fillna(0)

df["LoanDuration"] = df["ProductCreditName"].map({
    "Cầm Cố Điện Thoại": 6,
    "Cầm Cố Xe Máy": 12,
    "Vay Trực Tuyến Qua Sim": 3
}).fillna(6)

def phan_loai_diem_tin_dung(score):
    if score < 500:
        return "Low"
    elif 500 <= score <= 700:
        return "Medium"
    else:
        return "High"

df["PhanLoaiDiemTinDung"] = df["TS_CREDIT_SCORE_V2"].apply(phan_loai_diem_tin_dung)

job_credit_score_avg = df.groupby("JobName").agg({"TS_CREDIT_SCORE_V2": "mean"}).reset_index()

gender_bad_debt_total = df.groupby("Gender").agg({"HasBadDebt": "sum"}).reset_index()

status_score_group = df.groupby(["Trạng thái", "PhanLoaiDiemTinDung"]).size().reset_index(name="Count")

df
```

# Xử lý dữ liệu chuỗi

```python
import re

def clean_string(text):
    if pd.isna(text):
        return text
    text = re.sub(r'[^\w\s]', '', text)
    # Loại bỏ khoảng trắng thừa
    text = ' '.join(text.split())
    return text

df['Hình thức cư trú'] = df['Hình thức cư trú'].apply(clean_string)
df['ReceiveYourIncomeSalary'] = df['ReceiveYourIncomeSalary'].apply(clean_string)

df['Hình thức cư trú'] = df['Hình thức cư trú'].str.lower()
df['ReceiveYourIncomeSalary'] = df['ReceiveYourIncomeSalary'].str.lower()

df
```

# Mã hóa thuộc tính

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
import os

categorical_columns = [
    'Gender', 'Hình thức cư trú', 'Thời gian đã sống', 'JobName',
    'ReceiveYourIncomeSalary', 'ProductCreditName', 'InterestPaymentType', 'Trạng thái'
]

df['Gender'] = df['Gender'].astype(str)
df['Gender'] = df['Gender'].replace('nan', 'Unknown')

label_encoder = LabelEncoder()
df['JobName_LabelEncoded'] = df['JobName'].fillna('Unknown')
df['JobName_LabelEncoded'] = label_encoder.fit_transform(df['JobName_LabelEncoded'])

onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
onehot_columns = ['Gender', 'ReceiveYourIncomeSalary']

df[onehot_columns] = df[onehot_columns].fillna('Unknown')
onehot_encoded = onehot_encoder.fit_transform(df[onehot_columns])

onehot_feature_names = onehot_encoder.get_feature_names_out(onehot_columns)

onehot_df = pd.DataFrame(onehot_encoded, columns=onehot_feature_names, index=df.index)

df = pd.concat([df, onehot_df], axis=1)

residence_order = [
    'Thuê', 'Thuộc sở hữu của họ hàng', 'Đồng sở hữu', 'Sở hữu cá nhân', 'Nhà sở hữu'
]
thoi_gian_da_song = [
    'Dưới 6 tháng', '6 tháng - 1 năm', '1 - 3 năm', '3 - 5 năm',
    '5 - 10 năm', 'Hơn 10 năm', '-1'
]
ordinal_encoder = OrdinalEncoder(categories=[residence_order, thoi_gian_da_song], handle_unknown='use_encoded_value', unknown_value=-1)
ordinal_columns = ['Hình thức cư trú', 'Thời gian đã sống']

df[ordinal_columns] = df[ordinal_columns].fillna('Unknown')
ordinal_encoded = ordinal_encoder.fit_transform(df[ordinal_columns])

ordinal_df = pd.DataFrame(ordinal_encoded, columns=['Residence_Ordinal', 'thoi_gian_da_song'], index=df.index)

df = pd.concat([df, ordinal_df], axis=1)

folder_path = '/content/ggdrive/MyDrive/Đề án mô hình học máy dự đoán nợ xấu/'

output_file = os.path.join(folder_path, 'processed_data.csv')
df.to_csv(output_file, index=False, encoding='utf-8')

df

```
