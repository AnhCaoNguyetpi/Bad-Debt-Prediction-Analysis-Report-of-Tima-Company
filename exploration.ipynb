  Vẽ biểu đồ Histogram để kiểm tra phân phối của số tiền vay ban đầu, tiền giải ngân và lương.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style= 'whitegrid', palette= 'pastel')

plt.figure(figsize=(15, 10))

plt.subplot(1,3,1)
sns.histplot(df['SoTienDKVayBanDau'], kde = True,bins = 30,color = 'blue')
plt.title('Phân phối số tiền vay đăng kí ban đầu')
plt.xlabel('SoTienVayDKBanDau')

plt.subplot(1,3,2)
sns.histplot(df['TienGiaiNgan'],kde = True, bins = 30, color='orange')
plt.title('Phân phối tiền giải ngân')
plt.xlabel('TienGiaiNgan')

plt.subplot(1,3,3)
sns.histplot(df['Salary'], kde = True,bins = 30,color = 'green')
plt.title('Phân phối thu nhập')
plt.xlabel('Salary')
```

  Vẽ Boxplot để phân tích sự phân tán và các giá trị ngoại lai trong SoTienDKVayBanDau, TienGiaiNgan và thu nhập.

```python
plt.figure(figsize = (8,4))
sns.boxplot(x = df['SoTienDKVayBanDau'],color='blue')
plt.title('Phát hiện giá trị ngoại lai: Số tiền ĐK ban đầu')
plt.xlabel('SoTienDKVayBanDau')

plt.figure(figsize = (8,4))
sns.boxplot(x = df['TienGiaiNgan'],color='green')
plt.title('Phát hiện giá trị ngoại lai: Số tiền giải ngân')
plt.xlabel('TienGiaiNgan')

plt.figure(figsize = (8,4))
sns.boxplot(x = df['Salary'],color='orange')
plt.title('Phát hiện giá trị ngoại lai: thu nhập')
plt.xlabel('Salary')
```

Vẽ Scatter plot giữa SoTienDKVayBanDau và TienGiaiNgan để kiểm tra mối quan hệ giữa số tiền vay và số tiền giải ngân.

```python
plt.figure(figsize = (8,4))
sns.scatterplot(x = 'SoTienDKVayBanDau', y='TienGiaiNgan',hue ='Gender',data = df,alpha = 0.7)
plt.title('Mối quan hệ giữa SoTienDKVayBanDau và TienGiaiNgan')
plt.xlabel('SoTienDKVayBanDau')
plt.ylabel('TienGiaiNgan')
plt.legend(title='Gender', loc='upper left')
```

Vẽ Scatter plot giữa Salary và SoTienDKVayBanDau để phân tích sự tương quan giữa thu nhập và khoản vay.

```python
plt.figure(figsize = (8,4))
sns.scatterplot(x = 'SoTienDKVayBanDau', y='Salary',hue ='Trạng thái',data = df,alpha = 0.7)
plt.title('Mối quan hệ giữa SoTienDKVayBanDau và Salary')
plt.xlabel('SoTienDKVayBanDau')
plt.ylabel('Salary')
plt.legend(title='Trạng thái', loc='upper left')
```

Vẽ Countplot cho Gender,Trạng Thái,CityName,JobName để phân tích sự phân bố trong bộ dữ liệu.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Đặt kích thước chung cho các biểu đồ
plt.figure(figsize=(15, 10))

# Vẽ countplot cho Gender
plt.subplot(2, 2, 1)
sns.countplot(x='Gender', data=df, hue='Gender', dodge=False)
plt.title('Phân bố theo Gender')
plt.xlabel('Gender')
plt.ylabel('Số lượng')

# Vẽ countplot cho Trạng Thái
plt.subplot(2, 2, 2)
sns.countplot(x='Trạng thái', data=df, hue='Trạng thái', dodge=False)
plt.title('Phân bố theo Trạng Thái')
plt.xlabel('Trạng Thái')
plt.ylabel('Số lượng')

# Vẽ countplot cho CityName
plt.subplot(2, 2, 3)
sns.countplot(y='CityName', data=df, order=df['CityName'].value_counts().index, hue = None)
plt.title('Phân bố theo CityName')
plt.xlabel('Số lượng')
plt.ylabel('CityName')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

Vẽ Heatmap để kiểm tra mối quan hệ giữa các biến liên tục như SoTienDKVayBanDau, TienGiaiNgan, Salary.

```python
cols = ['SoTienDKVayBanDau','TienGiaiNgan','Salary']

ma_tran = df[cols].corr()

plt.figure(figsize = (6,4))

sns.heatmap(ma_tran,annot = True,fmt='.2f',cmap ='coolwarm', cbar = True)

plt.title('Mối quan hệ tương quan giữa các biến')
plt.xticks(rotation = 45,fontsize =8)
plt.yticks(rotation = 45,fontsize = 8)

plt.show()
```

Vẽ Heatmap để phân tích sự tương quan giữa các biến phân loại như Gender, CityName, và các yếu tố tài chính.

```python
cot_bien_phan_loai = ['Gender']
cot_yeu_to_tai_chinh = ['SoTienDKVayBanDau','TienGiaiNgan']

df_copy = df.copy()

df_copy = pd.get_dummies(df_copy, columns=cot_bien_phan_loai, drop_first=True)

selected_cols = [col for col in df_copy.columns if col.startswith('Gender') or col in cot_yeu_to_tai_chinh]

ma_tran = df_copy[selected_cols].corr()

plt.figure(figsize = (10,8))
sns.heatmap(
    ma_tran,annot = True, fmt = ".2f", cmap = "YlGnBu", linewidths = 0.5, cbar = True
)

plt.title('Heatmap: Tương quan giữa các biến phân loại và tài chính', fontsize=14)
plt.xticks( rotation=45, fontsize = 10)
plt.yticks( rotation=0, fontsize = 10)
```

Vẽ Pairplot giữa các biến như SoTienDKVayBanDau, TienGiaiNgan, Salary, và NumberOfLoans để khám phá các mối quan hệ giữa các biến.

```python
cols = ['SoTienDKVayBanDau','TienGiaiNgan','Salary','NumberOfLoans']

sns.pairplot(df[cols],diag_kind = 'kde',kind = 'scatter', plot_kws={'alpha': 0.6})

plt.suptitle('Pairplot: Mối quan hệ giữa các biến liên tục', y=1.02, fontsize=16)
plt.show()
```

Vẽ Bubble chart để kiểm tra mối quan hệ giữa SoTienDKVayBanDau, TienGiaiNgan và Salary, với kích thước bong bóng thể hiện NumberOfLoans.

```python
import numpy as np
df['NumberOfLoans'] = df['NumberOfLoans'].fillna(0)
df['NumberOfLoans'] = np.abs(df['NumberOfLoans'])
plt.figure(figsize=(6,4))

bubble = plt.scatter( df['SoTienDKVayBanDau'],
    df['TienGiaiNgan'],
    s=df['NumberOfLoans'] * 50,
    c=df['Salary'],
    cmap='viridis',
    alpha=0.7,
    edgecolors='w'
)

plt.colorbar(bubble, label='Salary')

plt.title('Bubble Chart: Mối quan hệ giữa SoTienDKVayBanDau, TienGiaiNgan và Salary', fontsize=14)
plt.xlabel('SoTienDKVayBanDau (Số tiền đăng ký vay ban đầu)', fontsize=12)
plt.ylabel('TienGiaiNgan (Tiền giải ngân)', fontsize=12)

plt.grid(alpha=0.3)
plt.show()
```

Vẽ Stacked bar chart để phân tích số lượng các khoản vay trong các nhóm phân loại như Gender, HasBadDebt và HasLatePayment.

```python
stacked_data = df.groupby(['Gender', 'HasBadDebt', 'HasLatePayment']).size().unstack(fill_value=0)

stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Phân tích số lượng khoản vay theo các nhóm phân loại', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Số lượng khoản vay', fontsize=12)
plt.legend(title='HasBadDebt & HasLatePayment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Vẽ Bar Plot để so sánh Salary giữa các nhóm Gender
